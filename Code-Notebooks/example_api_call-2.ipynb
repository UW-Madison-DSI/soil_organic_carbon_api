## Soil Organic Carbon XGBoost Model

Our API operationalizes the modeling behind the soc prediction in conus

```python
import requests

url = 'https://connect.doit.wisc.edu/soil_organic_carbon_prediction/v1/prediction'

headers = {
    'accept': 'application/json',
    'Content-Type': 'application/json'
}

data = {
    "depth_cm": 5,
    "total_precipitation": 665.71,
    "mean_temperature": 19.702397,
    "dem": 1646.838135,
    "slope": 1.61754,
    "aspect": 25.694852,
    "hillshade": 178.0,
    "bd_mean": 1.317050,
    "clay_mean": 15.225413,
    "om_mean": 0.063926,
    "ph_mean": 7.000336,
    "sand_mean": 55.055664,
    "land_use": 5,
    "land_cover": 5
}

response = requests.post(url, headers=headers, json=data)

if response.status_code == 200:
    print("Prediction successful!")
    print(response.json())
else:
    print(f"Failed to get prediction: {response.status_code}")
    print(response.text)
```
